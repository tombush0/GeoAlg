```python
def triangul(lines):
    
    poly = get_points(lines)
    
    if not monoton(lines):
        return []
    
    #top, vertice, reshaping list of vertices to have top at 0
    p_max_y = sorted(poly, key = lambda x: x[1], reverse = True)[0]
    y_max_index = poly.index(p_max_y)
    poly = poly[y_max_index:] + poly[:y_max_index]
    #y_max_index = 0
    
#     print("Points - start")
#     for p in poly:
#         print(p)
#     print('------')
    
    
    #structure of points: position and chain
    class v:
        def __init__(self,crd, left):
            self.pos = crd
            self.left = left
            
    polygon = [v(poly[0], True)]
    #polygon.append(v(poly[0], True))
    
    #chain splitting
    for p in range(1, len(poly)):
        if poly[p][1] < poly[p-1][1]:
            polygon.append(v(poly[p], True))
        else:
            polygon.append(v(poly[p], False))
    
#     print("Points to chains")
#     for p in polygon:
#         print(p.pos, p.left)
#     print('----')
    
    polygon = sorted(polygon, key = lambda p: p.pos[1], reverse = True)
    
#     print("Points sorted")
#     for p in polygon:
#         print(p.pos, p.left)
#     print('----')
            
    print("Stack - start")
    def print_stack(S):
        if len(S) == 0:
            print('Empty')
        else:
            for s in reversed(S):
                print(s.pos)
        print()

    
    diagonals = []
    scenes = [Scene([PointsCollection([])], [LinesCollection(lines)])]
    
    S = []
    S.append(polygon[0])
    S.append(polygon[1])
    
    for j in range(2, len(polygon)):
        print("Next vertice")
        print_stack(S)
        p = polygon[j]
        if p.left != S[-1].left or p == polygon[-1]:
            print("OPPOSITE CHAIN ")
            for i in range(len(S)-1):
                u = S.pop()
                
                print("Adding diagonal")
                #scenes.append( Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection(diagonals, color = "yellow")]) )
                scenes += [Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection([[p.pos, u.pos]], color = "red")])]
                #scenes += [Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection(diagonals, color = "yellow"), LinesCollection([[p.pos, u.pos]], color = "red")])]
                diagonals.append([p.pos, u.pos])
                print_stack(S)
            u = S.pop()
            S.append(polygon[j-1])
            S.append(polygon[j])
        else:
            print("SAME CHAIN ")
            #left/right chain
            if p.left == True:
                e = -1
            else:
                e = 1
                
            flag = True
            while len(S) > 1 and flag:
                u = S.pop()
                u_p = S.pop()
                print_stack(S)
                if det(p.pos, u.pos, u_p.pos)*e < 0:      #cutting triangle off
                    print("Adding diagonal")
                    #scenes.append( Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection(diagonals, color = "yellow")]) )
                    #scenes += [Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection(diagonals, color = "yellow"), LinesCollection([[p.pos, u.pos]], color = "red")])]
                    scenes += [Scene([PointsCollection([])], [LinesCollection(lines), LinesCollection([[p.pos, u.pos]], color = "red")])]
                    diagonals.append([p.pos, u_p.pos])
                    S.append(u_p)
                    u = u_p
                else:
                    print("Moving on")
                    to_stack = [p,u,u_p]
                    S.extend(to_stack)
                    flag = False
                print_stack(S)
            if flag:
                S.append(p)   
    
    return scenes
    
```
